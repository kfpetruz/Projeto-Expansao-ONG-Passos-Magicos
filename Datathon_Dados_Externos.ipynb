```python
import csv
# Caminho do arquivo CSV onde você deseja escrever os dados
caminho_arquivo = 'dados_coordenada_cidade.csv'

# Escrever os dados no arquivo CSV
with open(caminho_arquivo, mode='w', newline='') as arquivo_csv:
    escritor_csv = csv.writer(arquivo_csv, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for linha in coordenadas:
        escritor_csv.writerow(linha)
```

```python
from geopy.distance import geodesic

# Coordenadas geográficas de São Paulo (latitude e longitude)
sao_paulo_coord = (-23.5505, -46.6333)

# Dicionário com as coordenadas geográficas de outras cidades (substitua pelas coordenadas das cidades desejadas)
cidades = {
    'São Paulo': (-23.5505, -46.6333),
    'Rio de Janeiro': (-22.9068, -43.1729),
    'Belo Horizonte': (-19.9167, -43.9333),
    'Curitiba': (-25.4284, -49.2733),
    'Porto Alegre': (-30.0331, -51.2300)
}

# Calcular a distância entre São Paulo e cada cidade
for cidade, coord in cidades.items():
    distancia = geodesic(sao_paulo_coord, coord).kilometers
    print(f'Distância entre São Paulo e {cidade}: {distancia:.2f} km')
```

```python
import googlemaps

# Chave da API do Google Maps (substitua 'SUA_CHAVE_AQUI' pela sua chave real)
api_key = 'SUA_CHAVE_AQUI'

# Inicializar o cliente da API do Google Maps
gmaps = googlemaps.Client(key=api_key)

# Coordenadas geográficas de São Paulo
sao_paulo_coord = (-23.5505, -46.6333)

# Lista de cidades e seus nomes
cidades = [
    ('Rio de Janeiro', -22.9068, -43.1729),
    ('Belo Horizonte', -19.9167, -43.9333),
    ('Curitiba', -25.4284, -49.2733),
    ('Porto Alegre', -30.0331, -51.2300)
]

# Calcular a distância entre São Paulo e cada cidade
for cidade, lat, lng in cidades:
    # Calcular a distância da rota entre São Paulo e a cidade
    rota = gmaps.directions(sao_paulo_coord, (lat, lng), mode="driving", region="BR")
    distancia = rota[0]['legs'][0]['distance']['text']
    print(f'Distância entre São Paulo e {cidade}: {distancia}')
```
